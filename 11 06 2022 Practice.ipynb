```python
#assinging the data frame to the new data frame with the cabin removed
df=df.drop("Cabin", axis=1)
```

```python
df.shape
```

```python
average_value= df['Age'].mean()
```

```python
print(average_value)
```

```python
#replacing null values in age with the average of age
df['Age'].fillna(average_value, inplace=True)
```

```python
#checking for null values again
df.isnull().sum()
```

```python
#we still have some missing values on Embarked. So we are going to get the mode of Embarked and replace the null values with the mode
```

```python
mode_value= df['Embarked'].mode()[0]
```

```python
print(mode_value)
```

```python
#checking why we have an S instead of a value
df.info()
```

```python
#we can see that Embarked is an object.
```

```python
#replacing null values in Embarked with the mode of Embarked
df['Embarked'].fillna(mode_value, inplace=True)
```

```python
#checking for null values again
df.isnull().sum()
```
